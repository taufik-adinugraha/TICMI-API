```python
import pandas as pd
from TicmiAPI import *
```

```python
api = TICMIDataAPI(api_key='XXXXXXXX')
startDate = "2024-06-01"
endDate = "2024-06-28"
stock = "BBCA"
```

```python
# Trading data for a specific stock
data = api.trading(startDate, endDate, stock)
pd.DataFrame(data['data'])
```

```python
# Index weight data
data = api.index_weight(startDate, endDate, indexCode="LQ45")
pd.DataFrame(data['data']).sort_values(['Date', 'SecCode']).reset_index(drop=True)
```

```python
# Index data
data = api.index(startDate, endDate, indexCode="LQ45")
pd.DataFrame(data['data']).sort_values('Date').reset_index(drop=True)
```

```python
# Request corporate action for a specific stock
data = api.corporate_action(startDate, endDate, stock, action_type='rups')
data
```

```python
# Request news for a specific stock
data = api.stock_news(startDate, endDate, stock)
data
```
